```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load the training and testing datasets

train_df = pd.read_csv('C:/Users/xpert/Downloads/churn-bigml-80.csv')
test_df = pd.read_csv('C:/Users/xpert/Downloads/churn-bigml-20.csv')

print(train_df.head())
```

```python
print(test_df.head())
```

```python
# Quick validation

print("Missing values per column:\n", train_df.isnull().sum())
print(train_df.duplicated().sum())
```

```python
print("Missing values per column:\n", test_df.isnull().sum())
print(test_df.duplicated().sum())
```

```python
# Predict Total day charge from Total day minutes

x_train = train_df[['Total day minutes']]
y_train = train_df['Total day charge']

x_test = test_df[['Total day minutes']]
y_test = test_df['Total day charge']
```

```python
# Create and Fit the model

model = LinearRegression()
model.fit(x_train, y_train)

# Predict on the test set

y_pred = model.predict(x_test)
```

```python
# Coefficients

print("Intercept:", model.intercept_)
print("Slope:", model.coef_[0])
```

Interpretation:  
The intercept (0.0006) represents the baseline Total Day Charge when Total Day Minutes is 0.  
The slope (0.17) means that for every 1-minute increase in Total Day Minutes, the Total Day Charge increases by approximately 0.17 units.


```python
# Evaluation metrics

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

mse, r2
```

Evaluation: 
The model achieved an R-square of 0.9999999, indicating that almost all the variation in Total Day Charge is explained by Total Day Minutes.  
The mean squared error 8.24e-6 is extremely low, showing very accurate predictions.
